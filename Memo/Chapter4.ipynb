# 행렬-벡터 곱셈의 기하학적 변환

```python
import numpy as np
import matplotlib.pyplot as plt

# 원점 (벡터 시작점)
origin = np.array([[0, 0], [0, 0]])  # 두 개의 벡터를 위해 원점 정의

# 행렬 M과 벡터 x 정의
M = np.array([[2, 3], [2, 1]])
x = np.array([[1], [1.5]])  # 2x1 벡터
v = np.array([[1.5], [1]])  # 2x1 벡터
Mx = M @ x  # 변환된 벡터
Mv = M @ v
print(Mx)
print(Mv)

# 변환 전과 후의 벡터들을 하나의 배열로 정리
vectors = np.hstack((x, Mx))  # [ [x1, Mx1], [x2, Mx2] ]

# 그래프 설정
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 첫 번째 그래프: 변환 전 벡터
ax1.quiver(*origin, *x, angles='xy', scale_units='xy', scale=1, color=['r'], label='Original x')
ax1.set_xlim(-4, 4)
ax1.set_ylim(-4, 4)
ax1.set_xticks(range(-4, 5))
ax1.set_yticks(range(-4, 5))
ax1.axhline(y=0, color='black', linewidth=0.5)
ax1.axvline(x=0, color='black', linewidth=0.5)
ax1.grid(True)
ax1.set_title("Original Vector x")
ax1.legend()

# 두 번째 그래프: 변환 후 벡터
ax2.quiver(*origin, *vectors, angles='xy', scale_units='xy', scale=1, color=['r', 'b'], label=['Original x', 'Mx'])
ax2.set_xlim(-4, 6)
ax2.set_ylim(-4, 6)
ax2.set_xticks(range(-4, 7))
ax2.set_yticks(range(-4, 7))
ax2.axhline(y=0, color='black', linewidth=0.5)
ax2.axvline(x=0, color='black', linewidth=0.5)
ax2.grid(True)
ax2.set_title("Transformation: Mx")
ax2.legend()

plt.show()
```
